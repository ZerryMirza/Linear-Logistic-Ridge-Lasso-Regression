# Logistic Regression


Used for classification problems. The algorithm uses the concept of Linear Regression with a change that the output is a sigmoid function of the Linear Equation. 

z = beta*x + intercept
p = 1 / (1 + e^-z)

If we try to solve the classification with simple linear regression equation. The line & slope keeps changing for every new value in the equation. Because of which a threshold cannot be set on the regression line (probability) to classify the 2 classes. For example, for the regression line (probability) the threshold is .5 

=> for >= .5 the class is positive

=> for < .5 the class is negative


For a new value in positive class (say an outlier, away from the other points), a new regression line will be fitted to the problem, changing the slope and intercept, hence (might be) .5 threshold will give different results. Hence we need to change the output (probability) of regression line to a different function that fits the new point easily.



Keeping the above problem in mind, a sigmoid function is used to change the class value between 0 & 1

z = beta*x + intercept

p = 1 / (1+e^-z)

=> 1+e^-z = 1/p

=> e^-z = (1/p) - 1

=> e^-z = (1 - p)/p

=> 1 / e^z = (1-p)/p

=> e^z = p/(1-p)

=> log(e^z) = log( p/(1-p) )

=> z = log ( p/(1-p) )

=> This equation is called the Odds ratio. i.e. the output is the ratio of positive class / other class




 The cost function of Logistic Regression is also different than linear regression. 
 
 In Linear Regression, the cost function is:

=> J = sigma(y_hat - y)^2 / n
 
 
 

In Logistic Regression, y is futher a function of z i.e. sigmoid of z is y_hat.

So we calculate Cost function as:

z = 1 / 1+ e^-z

J = Log Loss = sigma -( y*(log(p) + (1-y)*(log(1-p) ) ) / N


Considering the above log loss formula, The cost function graph for the positive and negative classes comes out to be as below:



So when the class y = 1, the right side of the equation becomes 0 and left side becomes active.

log loss = sigma -(y*log(y_hat) + (1-y)log(1-y_hat)

y_hat = 1 / 1+e^-z
** Note that the sigmoid returns a value between 0 and 1. These values will be the output of the log function above. 

When y=1 right side of the equation becomes 0 and the cost is calculated for class y = 1 & predicted values i.e. log(p) or log(y_hat) that creates the graph above. For example if the actual class = 1 and the sigmoid function returns 0.7 then the loss function is;

log loss = -(y*log(y_hat) + (1-y)*(log(1-y_hat)) )

J = - (1*log(0.7) + (1-1)*(log(1-0.7)) )

J = - (1*(-0.04575749056) )

J = 0.04575749056
