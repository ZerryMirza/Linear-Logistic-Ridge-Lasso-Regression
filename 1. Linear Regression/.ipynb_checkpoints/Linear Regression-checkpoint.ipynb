```python
def hypothesis(X, theta_0, theta_1):
    return theta_0, X*theta_1


def cost_function(X, y, theta_0, theta_1):
    m = len(X)
    summation = 0.0
    for i in range(m):
        summation += ((X[i]*theta_1+theta_0) - y[i])**2 # Summation of cost function i.e. (y_hat - y_actual)
    
    return summation / (2*m)


def gradient_decent(X, y, theta_0, theta_1, learning_rate):
    der_0 = 0
    der_1 = 0
    m = len(X)
    for i in range(m):
        der_0 += (theta_1 * X[i] + theta_0) - y[i] # Summation of der_0
        der_1 += ((theta_1 * X[i] + theta_0) -y[i])* X[i] # Summation of der_1
    theta_0 -= (1/m)*learning_rate*der_0
    theta_1 -= (1/m)*learning_rate*der_1
    return theta_0, theta_1


def training(X, y, theta_0, theta_1, learning_rate, iters):
    cost_history = [0]
    t0_history = [0]
    t1_history = [0]
    for i in range(iters):
        theta_0, theta_1 = gradient_decent(X, y, theta_0, theta_1, learning_rate)
        t0_history.append(theta_0)
        t1_history.append(theta_1)
        cost = cost_function(X, y, theta_0, theta_1)
        cost_history.append(cost)
        if i % 10 == 0:
            print(f'iter={i}, theta_0={theta_0}, theta_1={theta_1}, cost={cost}')
    return t0_history, t1_history, cost_history
```

```python
theta_0 = np.random.random()
theta_1 = np.random.random()
```

```python
t0_history, t1_history, cost_history = training(X.values, y.values, theta_0, theta_1, 0.0015, 6000)
```
